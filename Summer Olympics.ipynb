# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
import numpy as ny
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("summer.csv")
```

```python
print("Cities :  ",len(df['City'].unique()))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold=df[df['Medal']=='Gold']

gold_medals=gold.groupby(by='Sport')['Medal'].count().sort_values(ascending=False).reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x='Sport',y='Medal',data=gold_medals.head(5))
plt.xticks(size=15)
plt.yticks(size=15)
plt.xlabel('Sport',size=20)
plt.ylabel('Gold Medals',size=20)
plt.show()

```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
medals=df.groupby(by='Sport')['Medal'].count().sort_values(ascending=False).reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x='Sport',y='Medal',data=medals.head(5))
plt.xticks(size=15)
plt.yticks(size=15)
plt.xlabel('Sport',size=20)
plt.ylabel('Medals',size=20)
plt.show()

```

### 4. Which player has won most number of medals? (Top 5)

```python
player_medals=df.groupby(by='Athlete')['Medal'].count().sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,10))
sns.barplot(x='Athlete',y='Medal',data=player_medals.head())
plt.xticks(size=15)
plt.yticks(size=15)
plt.xlabel('Athlete',size=20)
plt.ylabel('Medals',size=20)
plt.show()
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
player_gold=df[df['Medal']=='Gold']

gold_medals=player_gold.groupby(by='Athlete')['Medal'].count().sort_values(ascending=False).reset_index()

plt.figure(figsize=(15,5))
sns.barplot(x='Athlete',y='Medal',data=gold_medals.head(5))
plt.xticks(size=15)
plt.yticks(size=15)
plt.xlabel('Athlete',size=20)
plt.ylabel('Gold Medals',size=20)
plt.show()
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for Event in df['Event'].unique():
    data.append([Event , len(df[df['Event']  == Event])])

pd.DataFrame(data,columns = ['Event','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'Event', y = 'freq', kind = 'bar', figsize = (8,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
gold=df[df['Medal']=='Gold']

if(gold['Gender']=='Women'):
    gold_medals=gold.groupby(by='Sport')['Medal'].count().sort_values(ascending=False).reset_index()
    plt.figure(figsize=(8,5))
    sns.barplot(x='Gender',y='Medal',data=gold_medals.head(5))
    plt.xticks(size=15)
    plt.yticks(size=15)
    plt.xlabel('Sport',size=20)
    plt.ylabel('Gold Medals',size=20)
    plt.show()
```
